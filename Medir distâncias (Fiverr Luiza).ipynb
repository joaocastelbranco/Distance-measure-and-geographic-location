```python
#Importar locais para tabela
df = pd.read_excel(r'C:\Users\joaoc\Desktop\Intensivo Python\Fiverr Luiza\Pontos distancias.xlsx')
numeracao = list(range(1,176))
Ponto1_list = df[1].tolist()
Ponto2_list = df[2].tolist()
Ponto3_list = df[3].tolist()
Ponto4_list = df[4].tolist()
print (Ponto4_list)
```

```python
Distancia_recta_1_2_list =  []
Distancia_carro_1_2_list =  []
Ponto1_coordenadas_list =  []
Ponto2_coordenadas_list =  []

#Calcular distancias entre Ponto 1 e Ponto 2
navegador = webdriver.Chrome(ChromeDriverManager().install())
navegador.get("http://www.distanciascidades.com/pesquisa/")
x=0
# Obter Distancias 1 e 2
for element in Ponto1_list:
    
        #escrever ponto 1
        time.sleep(1)
        navegador.find_element_by_xpath('//*[@id="fr"]').click()
        navegador.find_element_by_xpath('//*[@id="fr"]').send_keys(Ponto1_list[x])
    
        #escrever ponto 2
        time.sleep(1)
        navegador.find_element_by_xpath('//*[@id="to"]').click()
        navegador.find_element_by_xpath('//*[@id="to"]').send_keys(Ponto2_list[x])
    
        #pesquisar
        navegador.find_element_by_xpath('//*[@id="fq"]/span[2]').click()
        time.sleep(3)
        
        #Recolher dados e voltar ao início
        Distancia_recta_1_2 = navegador.find_element_by_xpath('//*[@id="d_an"]').text
        Distancia_carro_1_2 = navegador.find_element_by_xpath('//*[@id="d_mn"]').text
        Ponto1_coordenadas = navegador.find_element_by_xpath('//*[@id="lls1"]').text
        Ponto2_coordenadas = navegador.find_element_by_xpath('//*[@id="lls2"]').text
        
        navegador.find_element_by_xpath('//*[@id="nav"]/a').click()
        
        #Adicionar os valores as listas de distancias
        Distancia_recta_1_2_list.append(Distancia_recta_1_2)
        Distancia_carro_1_2_list.append(Distancia_carro_1_2)
        Ponto1_coordenadas_list.append(Ponto1_coordenadas)
        Ponto2_coordenadas_list.append(Ponto2_coordenadas)
        
        x=x+1
        if x > 173:
            navegador.quit()
            print ("distancias entre 1 e 2 feitas")
            break
    
    
    
        
```
